# Import File and Library

```python
#import library
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#import file
df = pd.read_csv('movies.csv')
df.head()
```

# Data Cleansing Step

```python
#check number of column and rows
df.shape
```

```python
#check datatype of each columns
df.dtypes
```

```python
#check missing value
df.isna().sum()
```

```python
#drop column Gross because most are missing values
df = df.drop('Gross', axis=1)
```

```python
#check to see Gross column are completely dropped
df.head()
```

```python
#Drop all rows that have missing value
df = df.dropna()
```

```python
#Check number of column and rows after drop missing value
df.shape
```

```python
#Drop duplicate rows
df = df.drop_duplicates(subset=['MOVIES'])
```

```python
#change VOTES colunn to Float 
df['VOTES'] = df['VOTES'].str.replace(',','')
df['VOTES'] = df['VOTES'].astype('float')
```

```python
#check type of VOTES column again
df['VOTES'].dtypes
```

```python
#clean GENRE column
df['GENRE'] = df['GENRE'].str.strip('\n')
df.head()
```

```python
#clean ONE-LINE column
df['ONE-LINE'] = df['ONE-LINE'].str.strip('\n')
df.head()
```

```python
#clean SRARS column
df['STARS'] = df['STARS'].str.replace('\n',' ')
df.head()
```

```python
#extracting DIRECTOR to new columns
df['DIRECTOR'] = df['STARS'].str.extract(r"Director:\s*([^|]*)")
df['DIRECTOR'].fillna("No Data",inplace = True)
```

```python
#remove Director from STARS columnn
df['STARS'] = df['STARS'].str.replace(r'.*Stars:', '', regex=True)
```

```python
#check result after clean STARS column
df.head()
```

```python
# Extract Start_Year and End_Year to new columns
df[['Start_Year', 'End_Year']] = df['YEAR'].str.extract(r'(\d{4})[–-]?(\d{4})?')
df['End_Year'].fillna(df['Start_Year'], inplace=True)
```

```python
#Change Start_Year and End_Year data type to integer
df['Start_Year'] = df['Start_Year'].astype(int)
df['End_Year'] = df['End_Year'].astype(int)
```

```python
#Drop YEAR column after we have Start_Year and End_Year columns
df.drop('YEAR', axis=1, inplace=True)
```

```python
#Check result
df.head()
```

# Analysis Step

```python
#Distribution of rating
plt.figure(figsize=(10, 6))
plt.hist(df['RATING'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Correlation heatmap
corr_col = df[['RATING','RunTime','VOTES']]
corr_matrix = corr_col.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Trend of Total Netflix Movies and Series After 2000
filter_years = df[df['Start_Year'] > 2000]
movies_year = filter_years['Start_Year'].value_counts()

plt.figure(figsize=(14, 6))
sns.lineplot(x=movies_year.index, y=movies_year.values, marker='o')
plt.title('Trend of Total Netflix Movies and Series After 2000')
plt.xlabel('Year')
plt.ylabel('Number of Released')
plt.xticks(ticks=movies_year.index, labels=movies_year.index, rotation=60, ha='right')
plt.grid(True)
plt.show()
```

#### We see that years below 2010 have a few movies in our data, so we don't use them to analyze average ratings.

```python
#Average Netflix Movies and Series Rating Per years
filter_years = df[df['Start_Year'] > 2010]
average_rating_per_year =filter_years.groupby('Start_Year')['RATING'].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=average_rating_per_year.index, y=average_rating_per_year.values, marker='o')
plt.title('Average Netflix Movies and Series Rating Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.xticks(ticks=average_rating_per_year.index, labels=average_rating_per_year.index, rotation=60, ha='right')
plt.grid(True)
plt.show()
```

#### The year with the highest average movie rating is 2018.
